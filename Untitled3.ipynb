```python
sales=([50,60,70,80,90,95,100,150,250,300])
import matplotlib.pyplot as plt
plt.hist(sales,color="green")
plt.title("sales peak")
plt.xlabel("len")
plt.ylabel("hig")
plt.show()
```

```python
profit=([10,20,50,60,80,60,70,80,90,95,100,150,250,30,95,150,200,201])
data=profit
data
```

```python
from scipy.stats import skew
a=skew(data)
if a>0:
    print("+ve")
elif a<0:
    print("-ve")
else:
    print("normal")
```

```python
Profit=([200,190,195,180,170,150,100,80,60,20,5,2,1])
Data=Profit
Data
```

```python
from scipy.stats import skew
a=skew(Data)
if a>0:
    print("+ve")
elif a<0:
    print("-ve")
else:
    print("normal")
```

```python
import matplotlib.pyplot as plt
plt.hist(Data,color="green")
plt.title("profit")
plt.xlabel("len")
plt.ylabel("hig")
plt.show()
```
